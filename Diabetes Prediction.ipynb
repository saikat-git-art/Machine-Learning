```python
'''

Assignment 5: Diabetes Prediction
Dataset: PIMA Diabetes
Link: https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv
Task:
 Predict diabetes outcome.
 Replace 0 values in features like BMI, Glucose with median.
 StandardScaler for all features.
 Train logistic regression and interpret coefficients
'''
```

```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
import numpy as np
```

```python
z = df[df['Glucose']==0]
print(z)
```

```python
df['Glucose'] = df['Glucose'].replace(0,df['Glucose'].median())
```

```python
df[df['BloodPressure']==0]
```

```python
df['BloodPressure'] = df['BloodPressure'].replace(0,df['BloodPressure'].median())
```

```python
df['SkinThickness'] = df['SkinThickness'].replace(0,df['SkinThickness'].median())
```

```python
df['Insulin'] = df['Insulin'].replace(0,df['Insulin'].median())
```

```python
df[df['BMI']==0]
```

```python
df['BMI'] = df['BMI'].replace(0,df['BMI'].median())
```

```python
df[df['DiabetesPedigreeFunction']==0]
```

```python
df[df['Age']==0]
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.iloc[:,0:8])
```

```python
df.iloc[:,0:8] = scaler.transform(df.iloc[:,0:8])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,0:8]
```

```python
y = df['Outcome'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
