```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/vega/vega-datasets/master/data/cars.json"
df = pd.read_json(url)
```

```python
df
```

```python
col = ['Miles_per_Gallon','Horsepower','Weight_in_lbs']
```

```python
df1 = df[col]
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
sns.boxplot(df1[col[0]])
```

```python
sns.distplot(df1['Miles_per_Gallon'])
```

```python
df1['Miles_per_Gallon'] = df1['Miles_per_Gallon'].fillna(df1['Miles_per_Gallon'].mean())
```

```python
sns.distplot(df1['Miles_per_Gallon'])
```

```python
sns.distplot(df1['Horsepower'])
```

```python
df1['Horsepower'].quantile(0.75)
```

```python
q3 = df1['Miles_per_Gallon'].quantile(0.75)
```

```python
IQR = (df1['Miles_per_Gallon'].quantile(0.75)) - (df1['Miles_per_Gallon'].quantile(0.25))
```

```python
df1['Horsepower'] = df1['Horsepower'].fillna(q3+1.5*IQR)
```

```python
sns.distplot(df1['Horsepower'])
```

```python
df1.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df2 = scaler.fit_transform(df1)
```

```python
from sklearn.cluster import KMeans
```

```python
error = []
for i in range(1,15):
    km = KMeans(n_clusters=i)
    km.fit(df2)
    error.append(km.inertia_)
print(error)
```

```python
plt.plot(range(1,15),error)
```

```python
km1 = KMeans(n_clusters=4)
```

```python
km1.fit_predict(df2)
```

```python
clusters = km1.fit_predict(df2)
```

```python
df3 = pd.DataFrame(df2,columns=df1.columns)
```

```python
df3
```

```python
sns.scatterplot(x=df3['Horsepower'],y=df3['Weight_in_lbs'],data=df3,hue=clusters,palette='deep')
```
