```python
'''
6. Height vs Weight Prediction
Dataset: https://raw.githubusercontent.com/selva86/datasets/master/HeightWeight.csv
Task: Predict weight based on height using Linear Regression and compare with
Polynomial Regression.

'''
```

```python
import pandas as pd
```

```python
df = pd.read_csv('heights_weights_sample.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['height'],df['weight'])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Height v/s Weight ')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df['height'].values
```

```python
x = x.reshape(-1,1)
```

```python
y = df['weight'].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,c='red')
```

```python
z = y_test - y_pred
```

```python
plt.scatter(y_pred,z)
```
